```python
# 1 + 2 + 3 ... + N
n = int(input("quantos números? "))
cont = 1
soma = 0
while cont <= n:
  valor = int(input("entre com um valor "))
  cont = cont + 1
  soma = soma + valor
print(f"A soma dos {n} primeiros inteiros é {soma}")
```

```python
# 1 + 2 + 3 ... + N
n = int(input("quantos números? "))
cont = 1
soma = 0
while cont <= n:
  soma += cont
  cont += 1
print(f"A soma dos {n} primeiros inteiros é {soma}")
```

```python
num = int(input("Enter a number: "))
while num != 0:
  if (num%2 == 0):
    print(f'{num} é par')
  else:
    print(f'{num} é impar')
  num = int(input('Insira um número: '))
print('Voce saiu do while!')
```

```python
# Programa para calcular o fatorial
num = int(input('Entre com um valor para calcular o fatorial: '))
if num < 0:
  print('Não existe fatorial de um número negativo...')
if num == 0:
  print("O fatorial de 0 é 1")
else:
  fat = 1
  i = 1
  while (i <= num):
    fat = fat * i
    i = i + 1
print(f'O fatorial do número digitado é {fat}')
```

```python
# Sequencia de fibonnaci
limite = 10
a, b = 0, 1
termos = 2

while termos < limite + 2:
  print(a, end = ' ')
  a, b = b, a
```

```python
# fatorial
num = int(input("Enter com um valor: "))
# check if the number is negative, positive or zero
if num < 0:
  print("Não existe fatorial de número negativo")
else:
  fat = 1
  for i in range(1, num+1, 1):
    fat = fat * i
  print(f"O fatorial de um {num} é {fat}")
```

```python
senha_correta = "123345"
tentativas = 3

while tentativas > 0:
  senha = input("Digite a senha: ")

  if senha == senha_correta:
    print("Senha correta. Acesso concedido.")
    break # Termina o loop
  else:
    tentativas -= 1
    print(f"Senha incorreta. Voce tem {tentativas} tentativas res")
print("Voce excedeu o número máximo de tentativas. Acesso bloqueado.")
```

```python
from ast import Continue
# Advinha o número
# Uma biblioteca é um conjunto de funcionaldiades, funções. Funcao é um conjunto de ferramentas.

import random
numero_secreto = random.randint(1, 10)
while True:
  palpite = int(input("Advinhe o número: "))

  if (palpite > numero_secreto):
    print("Valor alto")
    continue
  elif (palpite < numero_secreto):
    print("Valor baixo")
    continue
  elif (palpite == numero_secreto):
    print("Parabens, voce acertou!")
    break
  else:
    print("Tente inserir um valor compatível")
    continue
```
